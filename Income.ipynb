#### Task 6: Set poverty level of the members and the head of the house within a family.

```python
train1['Target'].value_counts()
```

```python
def poverty(train1):
    rating = []
    for key in train1['Target']:
        if key == 4:
            rating.append('Level 4')
        elif key == 3:
            rating.append('Level 3')
        elif key == 2:
            rating.append('Level 2')
        elif key == 1:
            rating.append('Level 1')
    return rating
```

```python
train1['poverty_level'] = poverty(train1)
```

###### Correlation

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```
